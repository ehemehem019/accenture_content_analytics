```python
import pandas as pd
import numpy as np
```

```python
content_copy = pd.read_csv('Content.csv')
reactions_copy = pd.read_csv('Reactions.csv')
reaction_types = pd.read_csv('ReactionTypes.csv')
```

```python
# Contents
# Delete rows with missing URLS using dropna()
content = content_copy.copy().dropna()
content.info()
```

```python
# Reactions
# Delete all rows with missing data
reactions = reactions_copy.copy().dropna()
reactions.info()
```

```python
# Reaction Types
# Nothing to delete
reaction_types.info()
```

```python
content.rename(columns = {"User ID": "Poster ID"}, inplace=True)
content
```

```python
reactions.rename(columns = {"User ID": "Audience ID"}, inplace=True)
reactions
```

```python
content_reactions = content.merge(reactions, how="left", on="Content ID")
content_reactions
```

```python
content_reactions.rename(columns = {"Type_x": "Content Type", "Type_y": "Emotion Type"}, inplace=True)
content_reactions
```

```python
content_reactions
```

```python
reaction_types.rename(columns = {"Type": "Emotion Type"}, inplace=True)
reaction_types
```

```python
content_reactions = content_reactions.merge(reaction_types, how="left", on="Emotion Type")
content_reactions
```

```python
content_reactions
```

```python
content_scoring = content_reactions[['Content ID', 'Poster ID', 'Content Type', 'Emotion Type', 'Datetime', 'Sentiment_x', 'Score_x']].copy()
content_scoring
```

```python
content_scoring.rename(columns = {"Sentiment_x": "Sentiment Type", "Score_x": "Score", "Poster ID": "Author ID"}, inplace=True)
content_scoring
```

```python
content_scoring.info()
# Convert Datetime to Date
# Convert Score to int
```

```python
content_scoring['Datetime'] = pd.to_datetime(content_scoring['Datetime'])
content_scoring.info()
```

```python
# Dataframe to excel
content_scoring.to_excel("C")
```

```python
# Top 5 Content
content_scoring.groupby(['Content ID', "Emotion Type"]).agg({'Score': 'sum'}).sort_values(by="Score", ascending=False).head(5)
```

```python
# Top emotions
content_scoring.groupby(['Content ID', "Content Type", "Emotion Type"])[['Score']].sum().head(5).sort_values(by="Score", ascending=False).head(5)
```
